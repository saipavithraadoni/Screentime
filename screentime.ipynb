```python
df['health'] = df['health'].map( 
                   {'Yes':1 ,'No':0}) 
```

```python
screen_time_chart = alt.Chart(df).mark_bar().encode(
    x='screen_time',
    y='count()',
    color='screen_time'
).properties(
    title='Screen Time Distribution'
)



screen_time_chart
```

```python
occupation_chart = alt.Chart(df).mark_bar().encode(
    x='occupation',
    y='count()',
    color='occupation'
).properties(
    title='Occupation Distribution'
)

occupation_chart 
```

```python
import altair as alt

# Data preparation might be needed for some visualizations
# For instance, converting categorical data to string if not already

# 1. Age Distribution
age_distribution_chart = alt.Chart(df).mark_bar().encode(
    alt.X("Age", bin=True),
    alt.Y('count()')
).properties(
    title='Age Distribution'
)

# 2. Screen Time Distribution by Age
screen_time_age_chart = alt.Chart(df).mark_bar().encode(
    alt.X('Age:O'),
    alt.Y('count()'),
    alt.Color('screen_time:N')
).properties(
    title='Screen Time Distribution by Age'
)

# 3. Device Usage by Occupation
device_occupation_chart = alt.Chart(df).mark_bar().encode(
    alt.X('occupation:N'),
    alt.Y('count()'),
    alt.Color('device:N'),
    alt.Column('device:N')
).properties(
    title='Device Usage by Occupation'
)

# 4. Health and Screen Time
health_screen_time_chart = alt.Chart(df).mark_bar().encode(
    alt.X('health:N'),
    alt.Y('count()'),
    alt.Color('screen_time:N')
).properties(
    title='Health and Screen Time'
)

# 5. Common Activities
# Preprocessing needed to make this chart. We need to count the occurrence of each activity.
activities_df = df['Activities'].str.split(', ', expand=True).stack().value_counts().reset_index()
activities_df.columns = ['Activity', 'Count']

activities_chart = alt.Chart(activities_df).mark_bar().encode(
    alt.X('Activity:N'),
    alt.Y('Count:Q')
).properties(
    title='Common Activities'
)

# Displaying all charts
(age_distribution_chart | screen_time_age_chart) & (device_occupation_chart | health_screen_time_chart) & activities_chart
```
